```python
#our imports
import matplotlib
import pyart
from copy import deepcopy
import netCDF4
from matplotlib import pyplot as plt
import numpy as np
from __future__ import print_function
import skfuzzy as fuzz
from scipy import ndimage
%matplotlib inline
```

```python
fname = '/data/PON11042015/PON-20150411-003006-PPIVol.nc'
```

```python
radar = pyart.io.read(fname)
```

```python
print(radar.fields.keys())
print(radar.nsweeps)
```

```python
radar.info('compact')
```

```python
print(radar.latitude['data'])
print(radar.longitude['data'])
print(radar.range['data'][-1]/1000.0)
```

```python
min_lon=108.
max_lon=111.
min_lat=-2
max_lat=.5
lon_lines=[107, 107.5, 108, 108.5, 109, 109.5,110,110.5,111,111.5]
lat_lines=[-1.5, -1,-0.5,0,.5,1,1.5,2]

#create an instance of the class using our radar
display = pyart.graph.RadarMapDisplay(radar)
#create a Matplotlib figure
f = plt.figure(figsize = [19,8])
#now we are going to do a three panel plot, resolution is a basemap parameter and determines the resolution of 
#the coastline.. here we set to intermediate or 'i' ('h' for high 'l' for low)
plt.subplot(1,2,1)
display.plot_ppi_map('DBZH', sweep = 1, vmin = -3, vmax = 60,
                     min_lon = min_lon, max_lon = max_lon, min_lat = min_lat, max_lat = max_lat,
                     lat_lines=lat_lines, lon_lines=lon_lines,
                     resolution = 'i')

plt.subplot(1,2,2)
display.plot_ppi_map('VELH', sweep = 1, vmin = -5, vmax = 5,
                     min_lon = min_lon, max_lon = max_lon, min_lat = min_lat, max_lat = max_lat,
                     lat_lines=lat_lines, lon_lines=lon_lines,
                     resolution = 'i')
```
